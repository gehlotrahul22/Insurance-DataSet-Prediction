# Importing Necessary Libraries

```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor

```

# Reading Data And Replacing Missing Values With 0

```python
data=pd.read_csv("train_qnU1GcL.csv")
data.head()
```

```python
data=pd.get_dummies(data)
data.head()
```

```python
X_train=data.drop("target",axis=1)
X_train.head()
```

```python
y_train=data["target"]
y_train.head()
```

```python
X_train.fillna(0,inplace=True)
```

```python
X_test=pd.read_csv("test_LxCaReE_Px2yuWs.csv")
X_test.head()
```

```python
X_test=pd.get_dummies(X_test)
X_test.head()
```

```python
X_test.fillna(0,inplace=True)
```

# Fitting Data Into Model And FInding Predictions

```python
# The mid model will use the default loss
mid_model = GradientBoostingRegressor(loss="ls")

# Fit models

mid_model.fit(X_train, y_train)

# Predict

pred= mid_model.predict(X_test)
```

# Placing Predictions In File

```python
testarr=list(X_test['id'])
data={'id':testarr,'target':pred}
pd.DataFrame(data).to_csv("final_submission_challenge.csv",index=False)
```
